```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 20, 30, 40, 50, 60, 70, 80])
mf = np.array([0.0, 0.3, 0.5, 0.8, 1.0, 0.8, 0.4, 0.2])
```

```python
centroid = fuzz.defuzz(x, mf, 'centroid')
bisector = fuzz.defuzz(x, mf, 'bisector')
mom = fuzz.defuzz(x, mf, 'mom')  
som = fuzz.defuzz(x, mf, 'som')  
lom = fuzz.defuzz(x, mf, 'lom')  
```

```python
print(f"Centroid: {centroid:.2f}")
print(f"Bisector: {bisector:.2f}")
print(f"Mean of Maximum (MoM): {mom:.2f}")
print(f"Smallest of Maximum (SoM): {som:.2f}")
print(f"Largest of Maximum (LoM): {lom:.2f}")
```

```python
plt.figure(figsize=(10, 7))
plt.plot(x, mf, label='Fuzzy Output', linewidth=2)
plt.fill_between(x, 0, mf, alpha=0.2)


plt.axvline(centroid, color='r', linestyle='--', label=f'Centroid: {centroid:.2f}')
plt.axvline(bisector, color='g', linestyle='--', label=f'Bisector: {bisector:.2f}')
plt.axvline(mom, color='b', linestyle='--', label=f'MoM: {mom:.2f}')
plt.axvline(som, color='m', linestyle='--', label=f'SoM: {som:.2f}')
plt.axvline(lom, color='c', linestyle='--', label=f'LoM: {lom:.2f}')

plt.title('Fuzzy Output and Defuzzified Result')
plt.xlabel('Output')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()
```
