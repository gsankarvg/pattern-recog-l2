4. Implement  a  method  to  compute  the  Euclidean  distance  between  two  grayscale 
images.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1 = cv2.imread('images/Flowers.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread('images/Road.png')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img1, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray')
```

```python
img1.shape
```

```python
img2.shape
```

```python
img1_flat = img1.flatten()
img2_flat = img2.flatten()
```

```python
diff = img1_flat - img2_flat
```

```python
squared_diff = diff ** 2
```

```python
sum_squares = squared_diff.sum()
```

```python
sq_root = sum_squares ** 0.5
```

```python
print(f"The euclidean distance between two grayscale images = {sq_root}")
```
